# Librerías

```python
## Para separacion del data
import shutil
```

# Separación del dataset de imágenes en train, test y validation

```python
# Esta es una salida de R con los IDs que van a test o val
# (los IDs ausentes irían a train). Los IDs funcionan así:
#   1 <= ID <= 1232 : kirmizi
#        ID >  1232 : siirt

file = open('pist_id.txt', 'r')
lines = file.readlines()

# En cada lista guardamos los IDs de cada conjunto,
# según la clase (kirmizi o siirt)
kTestIndex = []
sTestIndex = []
kValIndex = []
sValIndex = []


# Por cada línea del archivo...
for line in lines:

    # Leer IDs
    test = line.split('\t')[0] 
    val  = line.split('\t')[1]

    try:
        # Convertir a número entero
        test = int(test)
        val = int(val)
    except:
        # La primera línea es un encabezado (ignorar)
        continue

    # Asignar a cada semilla dependiendo del ID
    if test <= 1232:
        kTestIndex.append(test)
    else:
        sTestIndex.append(test - 1232)

    if val <= 1232:
        kValIndex.append(val)
    else:
        sValIndex.append(val - 1232)


# Al llegar hasta acá se tienen 4 listas con los IDs de
# las imágenes que van a los conjuntos test y val, para
# cada tipo de semilla

# A continuación se crean copias de las imágenes en los
# directorios correspondientes. Es necesario que las
# carpetas ya estén creadas:
#   - /img
#       - /test
#           - /kirmizi
#           - /siirt
#       - /train
#           - /kirmizi
#           - /siirt
#       - /val
#           - /kirmizi
#           - /siirt

# Se hace primero con kirmizi

origin = "Pistachio_Image_Dataset/Images/Kirmizi_Pistachio/"

for k in range(1, 1232+1):
    
    # A qué carpeta va a ir la imagen k-ésima
    if k in kTestIndex:
        partition = "test"
    elif k in kValIndex:
        partition = "val"
    else:
        partition = "train"

    destinationRoute = "./img/" + partition + '/kirmizi/'

    # Nombre de la copia (archivo nuevo)
    filename = str(k) + ".jpg"

    # Por algún motivo los nombres de las primeras 65 imágenes
    # tienen un formato distinto a las restantes, así que acá
    # se tiene en cuenta eso
    if k <= 65:
        originRoute = origin + "kirmizi (" + str(k) + ").jpg"
    else:
        originRoute = origin + "kirmizi " + str(k - 65) + ".jpg"

    # Se hace una copia de la imagen, llamada {filename}
    shutil.copy(originRoute, destinationRoute + filename)


# Y ahora lo mismo pero para siirt

origin = "Pistachio_Image_Dataset/Images/Siirt_Pistachio/"

for s in range(1, 916+1):

    if s in sTestIndex:
        partition = "test"
    elif s in sValIndex:
        partition = "val"
    else:
        partition = "train"

    destinationRoute = "./img/" + partition + '/siirt/'

    filename = str(s) + ".jpg"

    # Acá el cambio de formato en los nombres se da después
    # de las primeras 50 imágenes
    if s <= 50:
        originRoute = origin + "siirt (" + str(s) + ").jpg"
    else:
        originRoute = origin + "siirt " + str(s - 50) + ".jpg"

    shutil.copy(originRoute, destinationRoute + filename)
```
